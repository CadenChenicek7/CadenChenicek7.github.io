```python
from vega_datasets import data
import altair as alt
import pandas as pd
import ipywidgets as interact

df = pd.read_csv("data/Video_Games.csv")

df_clean = df.dropna()
df_clean.head()
```

```python
# here is the calculation for the weighted score
def get_total_score(row):
    #df1 = df.assign(Total_Score=lambda x: (x.User_Count/(x.User_Count+x.Critic_Count)) * int(x.User_Score)) 
    critic_score = row['Critic_Score']
    num_critic = row['Critic_Count']
    user_score = float(row['User_Score']) * 10
    num_user = row['User_Count']
    total = num_user + num_critic
    total_score = (float(num_user/total) * float(user_score)) + (float(num_critic/total) * float(critic_score))
    #df_clean['total_score'] = (df_clean['User_Count'] / (df_clean['User_Count'] + df_clean['Critic_Count'])) * float(df_clean['User_Score']) + (df_clean['Critic_Count'] / (df_clean['User_Count'] + df_clean['Critic_Count'])) * df_clean['Critic_Score']
    return total_score

# here is where i populate the df
def populate_df(df):
    dictionary = {
            "Name": [],
            "Year": [],
            "Platform": [],
            "Dev": [],
            "Genre": [],
            "Total_Score": [],
            "NA_Sales": [],
            "EU_Sales": [],
            "JP_Sales": [],
            "Other_Sales": [],
            "Global_Sales": []
            }
    for i in range(len(df)):
        dictionary["Name"].append(df.iloc[i]["Name"])
        dictionary["Year"].append(df.iloc[i]["Year_of_Release"])
        dictionary["Platform"].append(df.iloc[i]["Platform"])
        dictionary["Dev"].append(df.iloc[i]["Developer"])
        dictionary["Genre"].append(df.iloc[i]["Genre"])
        dictionary["Total_Score"].append(round(get_total_score(df.iloc[i]), 2))
        dictionary["NA_Sales"].append(df.iloc[i]["NA_Sales"])
        dictionary["EU_Sales"].append(df.iloc[i]["EU_Sales"])
        dictionary["JP_Sales"].append(df.iloc[i]["JP_Sales"])
        dictionary["Other_Sales"].append(df.iloc[i]["Other_Sales"])
        dictionary["Global_Sales"].append(df.iloc[i]["Global_Sales"])
        
    df_new = pd.DataFrame(dictionary)
    return df_new    

df_clean = populate_df(df_clean)
```

```python
source = df_clean.head(1000)

input_dropdown = alt.binding_select(options=['Nintendo', 'Rockstar North', 'Treyarch', 'Infinity Ward', 'Ubisoft', 'Bungie', 'Electronic Arts', 'The Sims Studio', 'Naughty Dog', 'Epic Games', 'Rare Ltd.', 'Capcom', 'Blizzard Entertainment', 'Valve Software', 'Konami', 'Bethesda Softworks', 'Maxis', 'EA Games', 'Rockstar Studios'], name='Developer: ')
selection = alt.selection_point(fields=['Dev'], bind=input_dropdown)
selection_interval = alt.selection_interval(bind='scales')
color = alt.condition(
    selection,
    alt.Color('Genre:N').scale(domain=['Action', 'Adventure', 'Fighting', 'Misc', 'Platform', 'Puzzle', 'Racing', 'Role-Playing', 'Shooter', 'Simulation', 'Sports', 'Strategy'],
                                range=['red', 'orange', 'magenta', 'navy', 'yellow', 'cyan', 'royalblue', 'limegreen', 'brown', 'purple', 'hotpink', 'seagreen']),
    alt.value('lightgray')
)

chart = alt.Chart(source, title='Top 1000 Best Selling Video Games by Rating and Sales').mark_point(opacity=0.6).encode(
    x = alt.X('Global_Sales:Q').title('Sales per Million'),
    y = alt.Y('Total_Score:Q').title('Score'),
    color=color,
    tooltip=['Name:N', 'Platform:N', 'Year:N']
).add_params(
    selection,
    selection_interval
).properties(
    width=900,
    height=300
)
chart.properties(width='container').save("video_games.json")
```
